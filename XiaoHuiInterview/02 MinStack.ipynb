```python
class MinStack():
    stack = []
    mins = []
    
    def __init__(self):
        self.stack = []
        self.mins = []
        
    def push(self, target):
        if self.mins:
            if target <= self.mins[-1]: # 等于也要添加
                self.mins.append(target)
        else:
            self.mins.append(target)
        self.stack.append(target)
    
    def pop(self):
        if self.stack:
            target = self.stack.pop()
            if self.mins and target == self.mins[-1]:
                self.mins.pop()
            return target
        else:
            return None
    
    def top(self):
        if self.stack:
            return self.stack[-1]
        else:
            return None
```

```python
stack = MinStack()
```

```python
stack.push(1)
stack.push(1)
stack.push(2)
stack.push(1)
stack.push(1)
```

```python
print("stack:",stack.stack)
print("mins:",stack.mins)
```

```python
stack.pop()
stack.pop()
stack.pop()
```

```python
print("stack:",stack.stack)
print("mins:",stack.mins)
```
